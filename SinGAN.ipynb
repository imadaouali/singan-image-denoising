```python
def PSNR(img1, img2, peak=255):
    '''
    Computes the PSNR between two images assumed to be in the range [0,1]
    '''
    x = ((np.array(img1).squeeze() - np.array(img2).squeeze()).flatten() )
    return (10*np.log10(peak**2 / np.mean(x**2)))
```

```python
img_clean = io.imread('Input/Images/book250.png').astype(np.float32)
img_clean = img_clean/255

img_noisy = io.imread('Input/Denoising/book_noisy_50.png').astype(np.float32)
img_noisy = img_noisy/255
```

```python
print('The PSNR between the clean image and the noisy image is: ', PSNR(img_clean, img_noisy, 1))
```

```python
list_of_files = glob.glob('Output/Denoising/book_different_alpha/*.png')
list_of_files.sort()
```

```python
for file in list_of_files:
  img_denoised = io.imread(file).astype(np.float32)
  img_denoised = img_denoised/255
  img_denoised = img_denoised[:,:,:-1]
  print(file)
  print(PSNR(img_clean, img_denoised, 1))
  print('-'*50)
```
